```python
from skimage import io ,transform ,feature,measure,filters,morphology
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import os
import numpy as np
import cv2
%matplotlib inline
%load_ext autoreload
%autoreload 2
import glob
from commonfunctions import *


```

```python
def deskew(img):
    
    #GET HOUGH TRANFROM
    hspace, angles, distances=transform.hough_line(img)
    
    #Find the top (angles,d) pairs that have the highest count in the Hough matrix 
    accum, angles_2, dists = transform.hough_line_peaks(hspace, angles, distances)
  
    
    #Calulate the skew angle as an average of the angles
    skewAngle = np.rad2deg(np.mean(angles_2))
   
    #Rotate the image by â€“ skew angle
    if(skewAngle<0):
        rotatedImg=transform.rotate(img, 90-abs(skewAngle),resize=True)
    else:
        rotatedImg=transform.rotate(img,skewAngle-90,resize=True)


        
    return rotatedImg
```

```python
def to_binary(img,val):
    threshold = np.copy(img)
    threshold[threshold<val]=0
    threshold[threshold>=val]=1
    return threshold
```

```python
def binarize(img):
    
    threshold = filters.threshold_otsu(img)
    img = to_binary(img,threshold)
    return img
```

```python
def segmentation(img):
    
    #apply closing 
    closedImage=morphology.binary_closing(1-img)
    
    #label the image
    label_image = measure.label(closedImage,connectivity=2,background=0)
    
    # show image after segmentation
    show_images([gaussianbinarized,label_image],['after binarization','label_image'])
    
    segmented_notes=[] #list of the segmented notes
    titles=[] #list of strings for image show titles
    i=1 #number of segment for image show titles
    
    #loop on each labeled segment
    for region in measure.regionprops(label_image):
        
        # take regions with large enough areas
        if region.area >= 50:
            
            # get rectangle around segmented notes
            minr, minc, maxr, maxc = region.bbox
            
            #take this part from original binarized image
            note=gaussianbinarized[minr:maxr+2,minc:maxc+2]
            
            #add it to the list
            segmented_notes.append(note)
            titles.append(str(i))
            i+=1

    
    return segmented_notes,titles
```

# Image Pre-Processing

```python
files = glob.glob ("./test-cases/16.jpg")
for file in files:
    img=io.imread(file)


    # Noise reduction
    gaussian = filters.gaussian(rgb2gray(img),sigma =0.6 )
    # Binarizing image using Otsu's Algorithm
    gaussianbinarized = binarize(gaussian)
    show_images([img,gaussian,gaussianbinarized],['image','After noise reduction','after binarization'])
```

```python
#deskweing Image 
deskewImage = deskew(1-gaussianbinarized)
show_images([gaussianbinarized,1-deskewImage],[' binarized image','after deskew'])
```
